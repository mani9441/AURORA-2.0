GROUP C â€” Apply BaselineModel (Anomaly Scoring)

```python
# Load trained baseline model
import joblib
import os

# 1. Define the path 
mine_id = "MINE_001"
model_dir = "baseline_models" 
model_path = os.path.join(model_dir, f"{mine_id}_baseline_model.joblib")

# 2. Load the entire bundle
model_bundle = joblib.load(model_path)

# 3. Extract the components
scaler = model_bundle["scaler"]
pca    = model_bundle["pca"]
FEATURES = list(model_bundle["features"])
#stats  = model_bundle["recon_error_stats"]
baseline_anamoly = model_bundle["baseline_anomaly"]


print(f"âœ… Successfully loaded model for {model_bundle['mine_id']}")
print(f"Features used: {FEATURES}")
```

```python
# Normalize using baseline statistics
X_mon = scaler.transform(monitor_df[FEATURES])
```

```python
# Project to baseline latent space

X_latent = pca.transform(X_mon)
```

```python
# Reconstruct and compute anomaly score

X_recon = pca.inverse_transform(X_latent)

anomaly_score = np.mean((X_mon - X_recon) ** 2, axis=1)
monitor_df["anomaly_score"] = anomaly_score
```

GROUP E â€” Convert Anomaly Scores Back to Raster (Per Date)

This is needed for:

Maps

Spatial logic

Visual validation

```python
import pandas as pd
```

```python
anomaly_maps = {}

for t in monitor_df["time"].unique():
    df_t = monitor_df[monitor_df["time"] == t]

    anomaly_img = (
        df_t
        .set_index(["y","x"])["anomaly_score"]
        .to_xarray()
    )

    anomaly_maps[pd.to_datetime(t)] = anomaly_img
```
